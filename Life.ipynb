```python
# Dummy encoding the categorical variables

dummies = pd.get_dummies(categorical, drop_first = True)
dummies.head()
```

```python
# Creating new object 'data1' to store the merged data

data1 = pd.concat([data, dummies],axis=1)

data1.head()
```

```python
data1.shape
```

```python
data1.drop(['Country','Status'],axis=1,inplace=True)
```

```python
data1.shape
```

```python
# Plotting correlation matrix

corrmat = data.corr()

plt.figure(figsize=(30, 15))

sns.heatmap(corrmat[(corrmat >= 0.9) | (corrmat <= -0.9)], 
            cmap='YlGnBu', vmax=1.0, vmin=-1.0,
            annot=True, annot_kws={"size": 15})

plt.title('Correlation between features', fontsize=15)
plt.show()
```

Here we can observe that the variables 'under five deaths' and 'infant deaths' are perfectly correlated.
The variiables 'thinness 5-9 years' and 'thinness 1-19 years' are highly correlated.
Hence we choose to drop 'under five deaths' and 'thinness 5-9 years'

```python
data1.drop(['under-five deaths ',' thinness 5-9 years'], axis=1,inplace=True)
```

```python
data1.shape
```

```python
# Handling missing data

data1.isnull().sum()
```

```python
# We had plotted the histogram for all the numeric variables
# We observed that all of them are heavily skewed.
# Hence we decide to fill the missing data for these variables with their Median
# BMI looks a bit evenly distributed.
# Hence we take it's mean
```

```python
data1['Life expectancy '] = data1['Life expectancy '].fillna(data1['Life expectancy '].median())

data1['Adult Mortality'] = data1['Adult Mortality'].fillna(data1['Adult Mortality'].median())

data1['Alcohol'] = data1['Alcohol'].fillna(data1['Alcohol'].median())

data1['Hepatitis B'] = data1['Hepatitis B'].fillna(data1['Hepatitis B'].median()) 

data1[' BMI '] = data1[' BMI '].fillna(data1[' BMI '].mean()) 

data1['Polio'] = data1['Polio'].fillna(data1['Polio'].median())

data1['Total expenditure'] = data1['Total expenditure'].fillna(data1['Total expenditure'].median())

data1['Diphtheria '] = data1['Diphtheria '].fillna(data1['Diphtheria '].median())

data1['GDP'] = data1['GDP'].fillna(data1['GDP'].median())

data1['Population'] = data1['Population'].fillna(data1['Population'].median())

data1[' thinness  1-19 years'] = data1[' thinness  1-19 years'].fillna(data1[' thinness  1-19 years'].median())

data1['Income composition of resources'] = data1['Income composition of resources'].fillna(data1['Income composition of resources'].median())

data1['Schooling'] = data1['Schooling'].fillna(data1['Schooling'].median())
```

```python
data1.isnull().sum()
```

# 1)- Linear Regression model without log of target variable

```python
# Importing sk.learn library and it's functions

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn import preprocessing
```

```python
# Importing statsmodels and its fucntions

import statsmodels
import statsmodels.api as sm
from statsmodels.compat import lzip
import statsmodels.stats.api as sms
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
from statsmodels.tools.eval_measures import rmse
```

```python
X = data1.drop('Life expectancy ', axis = 1)

y = data1['Life expectancy ']
```

```python
# Splitting ddata into training and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)

print("The shape of X_train is:",X_train.shape)

print("The shape of X_test is:",X_test.shape)

print("The shape of y_train is:",y_train.shape)

print("The shape of y_test is:",y_test.shape)
```

```python
# Fitting the OLS model

linreg_nolog_model = sm.OLS(y_train, X_train).fit()

# Print the summary output
print(linreg_nolog_model.summary())
```

```python
# Pedicting the model

linreg_nolog_model_predictions = linreg_nolog_model.predict(X_test)
```

```python
# Calculating rmse using rmse()
linreg_nolog_model_rmse = rmse(y_test, linreg_nolog_model_predictions)

# Calculating R-squared using rsquared
linreg_nolog_model_rsquared = linreg_nolog_model.rsquared

# Calculating Adjusted R-Squared using rsquared_adj
linreg_nolog_model_rsquared_adj = linreg_nolog_model.rsquared_adj 
```

```python
# Tabulating the results

linreg_nolog_model_metrics = pd.Series({'Model': "Lin reg without log of target variable ",
                                                 'RMSE':linreg_nolog_model_rmse,
                                                 'R-Squared': linreg_nolog_model_rsquared,
                                                 'Adj. R-Squared': linreg_nolog_model_rsquared_adj})


cols = ['Model', 'RMSE', 'R-Squared', 'Adj. R-Squared']

result = pd.DataFrame(columns = cols)

result = result.append(linreg_nolog_model_metrics, ignore_index = True)

result
```

# 2)- Linear Regression model with log of target variable

```python
log_life_expec = np.log(data1['Life expectancy '])

X = data1.drop('Life expectancy ', axis = 1)

y = log_life_expec
```

```python
# splitting the data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)

print("The shape of X_train is:",X_train.shape)

print("The shape of X_test is:",X_test.shape)

print("The shape of y_train is:",y_train.shape)

print("The shape of y_test is:",y_test.shape)
```

```python
# Fitting the model

linreg_log_model = sm.OLS(y_train, X_train).fit()

print(linreg_log_model.summary())
```

```python
# Fitting the model

linreg_log_model_predictions = linreg_log_model.predict(X_test)
```

```python
# Calculating rmse using rmse()
linreg_log_model_rmse = rmse(y_test, linreg_log_model_predictions)

# Calculating R-squared using rsquared
linreg_log_model_rsquared = linreg_log_model.rsquared

# Calculating Adjusted R-Squared using rsquared_adj
linreg_log_model_rsquared_adj = linreg_log_model.rsquared_adj
```

```python
linreg_log_model_metrics = pd.Series({'Model': "Lin reg with log of target variable ",
                                                 'RMSE':linreg_log_model_rmse,
                                                 'R-Squared': linreg_log_model_rsquared,
                                                 'Adj. R-Squared': linreg_log_model_rsquared_adj})


result = result.append(linreg_log_model_metrics, ignore_index = True)

result
```

# 3)- Random Forest Regressor

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
# Intantiating the regressor
rf_reg = RandomForestRegressor(n_estimators=100, random_state=10)

# Fitting the regressor with training dataset
rf_reg.fit(X_train, y_train)
```

```python
# Predicting the values on test dataset using predict()
y_pred = rf_reg.predict(X_test)
```

```python
# Calculating rmse using rmse()
rf_reg_RMSE = np.sqrt(metrics.mean_squared_error(y_test, y_pred))

# Calculating R-squared using rsquared
rf_reg_r2 = r2_score(y_test,y_pred)
```

```python
rf_model_metrics = pd.Series({'Model': "Random forrest regressor ",
                                                 'RMSE':rf_reg_RMSE,
                                                 'R-Squared': rf_reg_r2})


result = result.append(rf_model_metrics, ignore_index = True)

result
```

# Conclusion

For this regression problem, I used 2 models; Linear regression using Ordinary Least Squares(OLS) and Random Forest regressor.

In order to get a more optimal solution, i did a few tuning steps in preprocessing such as removing highly correlated variables.

In order to get a more thorough result, i took log of the target variable which did help in achieving a way better error rate.

I then used this log converted target variable to pass it through the random forest regressor which proved to be the best algortihm giving the least RMSE and best R-squared.

Sk.learn does not explicitely offer Adjusted-R for Random forest regressor.

So given that our RMSE is the lowest and R-squared highest, we can conclude that The Random Forest regressor is the the best model for 'Life expectancy Prediction'
